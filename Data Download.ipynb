```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

**search for dengue protien - Dengue virus type 2 NS3 protein**

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('dopamine')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#selecting the single protein
selected = targets.target_chembl_id[39]
selected 
#unique target id
```

#
Get the bioactivity data for this CHEMBL5980
This bioactivity data is reported as IC50 values in nM nanoMolar Units

```python
activity=new_client.activity
#res=activity.filter(target_chembl_id=selected).filter(standard_type="Activity")
res=activity.filter(target_chembl_id=selected)
```

```python
df=pd.DataFrame.from_dict(res)
```
